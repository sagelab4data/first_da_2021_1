## **4.붓꽃 데이터로 의사결정나무, 랜덤포래스트 수행해보기**

```python
# 4.1 붓꽃 데이터를 읽어들인다. 
iris = sns.load_dataset('iris')
iris
```

```python
# 4.2 붓꽃 데이터를 피터와 타깃으로 나누기
data = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].to_numpy()
target = iris['species'].to_numpy()
```

```python
# 4.3 붓꽃 데이터를 학습데이터와 검증데이터 나누기

from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)
```

```python
# 4.4 붓꽃 데이터 개수 확인
print(train_input.shape, test_input.shape)
```

```python
# 4.5 의사결정나무 알고리즘 읽어들임
from sklearn.tree import DecisionTreeClassifier

# 모델 선택 (단, 가지치기는 깊이 3까지만 수행)
dt = DecisionTreeClassifier(max_depth = 3, random_state=42)
# 학습 수행
dt.fit(train_input, train_target)
```

```python
# 4.6 붓꽃 DT 학습 결과의 평가

# 학습 데이터 결과
print(dt.score(train_input, train_target))
# 검증 데이터 결과
print(dt.score(test_input, test_target))
```

```python
# 4.7 붓꽃 DT 데이터를 보기 쉽게 전달인자를 추가 
# filled=True : 색상 채워서
# feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] : 조건문에 항목명 표시

plt.figure(figsize=(20,15))
plot_tree(dt,  filled=True, feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.show()
```

```python
# 4.8 붓꽃 DT 항목의 영향력
# 'sepal_length', 'sepal_width',
# 'petal_length', 'petal_width'
print(dt.feature_importances_)
```

```python
# 4.9 붓꽃  랜덤포래스트 수행 및 결과 검증을 위한 라이브러리 임포트
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
```

```python
# 4.10 붓꽃 랜덤포래스트 모델 수행

rf = RandomForestClassifier (n_jobs = -1,  random_state = 42)
ret = rf.fit( train_input, train_target  )
```

```python
# 4.11 붓꽃 랜덤포래스트 학습 결과의 평가

# 학습 데이터 결과
print(rf.score(train_input, train_target))
# 검증데이터 결과
print(rf.score(test_input, test_target))
```

```python
# 4.12  붓꽃 랜덤포래스트 모델에서 각 항목의 영향력
print (rf.feature_importances_)
```
