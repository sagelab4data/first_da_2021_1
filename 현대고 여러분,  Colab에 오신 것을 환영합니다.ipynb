## **3.데이터 과학**

코랩을 통해 파이썬의 풍부한 데이터 분석 라이브러리를 활용하여 데이터를 분석하고 시각화할 수 있습니다.

* 아래 코드 셀에서는 <strong>넘파이Numpy</strong>를 사용하여 임의의 데이터를 생성하고 <strong>매트플롯립matplotlib</strong>으로 이를 시각화합니다. 

엑셀 또는 CSV 등의 데이터를 구글 드라이브 G Drive 또는 깃허브 GitHub계정에서 코랩 노트(메모장)으로 가져올 수 있고, 직접 데이터를 만들어서 사용 할 수  있습니다. 

* 아래 예제에서는 직접 난수를 발생 시키고 이를 그래프로 표현해 봅니다. 셀을 클릭하고 실행해 봅시다.




```
# 숫자 연산을 다루는 numpy와 그래프를 그리는 matplotlib을 읽어 들입니다.
import numpy as np
import matplotlib.pyplot as plt

# numpy를 이용하여 난수 random value를 발생 시킵니다. 
ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

# matplotlib을 이용하여 그래프를 그립니다.
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

# 그래프에 제목을 만들고 출력합니다. 
plt.title("Sample Visualization 난수발생 시각화")
plt.show()
```
